```python
import requests
from bs4 import BeautifulSoup as soup
```

```python
page = requests.get('http://books.toscrape.com/')
bsobj = soup(page.content,'lxml')
```

```python
bsobj
```

```python
genres = []
genre_link = []
for genre in bsobj.findAll('div',{'class':'side_categories'}):
    for name in genre.findAll('li'):
        genres.append(name.a.text.strip())
        genre_link.append(name.a['href'])
```

```python
genres
```

```python
genre_link
```

```python
import pandas as pd
```

```python
d1 = {"Genres":genres,"Genre_Link":genre_link}
```

```python
df_genre = pd.DataFrame(d1)
df_genre = df_genre.drop([0])
df_genre
```

```python

ratings = []
title = []
price = []
link = []

for page in range(1,51):
    html = requests.get('http://books.toscrape.com/catalogue/page-{}.html'.format(page))
    bsobj = soup(html.content,'lxml')
    for book in bsobj.findAll('article',{'class':'product_pod'}):
        ratings.append(book.p['class'][-1])
        title.append(book.h3.text.strip())
        link.append(book.a['href'])
    for prices in bsobj.findAll('p',{'class':'price_color'}):
        price.append(float(prices.text.strip().replace('£','')))
```

```python
d2 = {'Book_Title':title,'Price(£)':price,'Ratings':ratings,'Books_Link':link}
```

```python
pd.set_option('display.max_colwidth', None)
df_books = pd.DataFrame(d2)
df_books.head()
```

```python
title = []
description = []
upc = []
ptype = []
p1 = []
p2 = []
tax = []
av = []
nr = []

for links in link:
    page = 'http://books.toscrape.com/catalogue/'+str(links)
    html = requests.get(page)
    bsobj = soup(html.content,'lxml')
    for book in bsobj.findAll('article',{'class':'product_page'}):
        title.append(book.div.h1.text.strip())
        price.append(float(book.div.p.text.strip().replace('£','')))
        description.append(bsobj.findAll('p')[3].text.strip())
   
        upc.append(bsobj.findAll('tr')[0].td.text.strip())
        ptype.append(bsobj.findAll('tr')[1].td.text.strip())
        p1.append(float(bsobj.findAll('tr')[2].td.text.strip().replace('£','')))
        p2.append(float(bsobj.findAll('tr')[3].td.text.strip().replace('£','')))
        tax.append(float(bsobj.findAll('tr')[4].td.text.strip().replace('£','')))
        av.append(bsobj.findAll('tr')[5].td.text.strip())
        nr.append(bsobj.findAll('tr')[6].td.text.strip())
        
    
```

```python
d3 = {'Title':title,'UPC':upc,'Product_Type':ptype,'Price (excl. tax)':p1,'Price (incl. tax)':p2,'Tax':tax,'Availability':av,'Number_of_reviews':nr,'Description':description}
```

```python
df_complete = pd.DataFrame(d3)
```
